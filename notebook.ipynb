## Content-Based Movie Recommendation Engine

This notebook walks through the full pipeline:
1. Data loading & exploration
2. Feature engineering (genres, keywords, cast, director, overview)
3. TF-IDF vectorisation
4. Cosine similarity computation
5. Recommendation generation
6. Evaluation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from recommender import (
    load_data, build_feature_soup,
    build_similarity_matrix,
    get_recommendations,
    evaluate_recommendations,
)

sns.set_theme(style='darkgrid')
pd.set_option('display.max_colwidth', 60)
```

## 1. Load & Explore Data

```python
df_raw = load_data('data/tmdb_5000_movies.csv', 'data/tmdb_5000_credits.csv')
print(f'Shape: {df_raw.shape}')
df_raw.head(3)
```

```python
# Rating distribution
plt.figure(figsize=(8, 4))
sns.histplot(df_raw['vote_average'], bins=20, kde=True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Average Rating')
plt.tight_layout()
plt.savefig('assets/rating_distribution.png', dpi=150)
plt.show()
```

## 2. Feature Engineering

```python
df = build_feature_soup(df_raw)
print('Sample soup:')
print(df['soup'].iloc[0][:300])
```

## 3. Build Similarity Matrix

```python
cosine_sim, indices = build_similarity_matrix(df)
print(f'Similarity matrix shape: {cosine_sim.shape}')
```

## 4. Generate Recommendations

```python
query = 'The Dark Knight'
recs  = get_recommendations(query, df, cosine_sim, indices, top_n=10)
print(f'Top 10 recommendations for "{query}":')
recs
```

```python
# Visualise similarity scores
plt.figure(figsize=(9, 4))
sns.barplot(data=recs, x='similarity_score', y='title', palette='mako')
plt.title(f'Similarity Scores – Top 10 for "{query}"')
plt.xlabel('Cosine Similarity')
plt.ylabel('')
plt.tight_layout()
plt.savefig('assets/similarity_scores.png', dpi=150)
plt.show()
```

## 5. Evaluation

```python
metrics = evaluate_recommendations(df, cosine_sim, indices, sample_size=200)
for k, v in metrics.items():
    print(f'{k:30s}: {v}')
```
